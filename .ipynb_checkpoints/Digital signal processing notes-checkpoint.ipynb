## A quick primer on Fourier Analysis

Most of DSP is based around Fourier Analysis, and so if anyone is not familiar with how they work it's worth doing a quick overview. I have a more elaborate set of notes specifically on these.

The basic idea of Fourier Analysis is this. If I have some continuous function $f(t)$ defined over some length $\left[0, T \right)$, I can generate function by adding a series of sines and cosines together of period $n/T$.

$$ f(t) = \sum_{0}^{\infty} a_0$$

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
np.random.seed(123)
t = np.linspace(-10, 10, 201)
x = 5*np.exp(-np.power(t + 2.3, 2))
x_noise = x + np.random.randn(len(t))
plt.plot(t, x, color='k')
plt.plot(t, x_noise, color='b')
```

```python
b, a = signal.butter(3, 0.05)
# mmm, butter
```

```python
zi = signal.lfilter_zi(b, a)
z, _ = signal.lfilter(b, a, x_noise, zi=zi*x_noise[0])
z2, _ = signal.lfilter(b, a, z, zi=zi*z[0])
y = signal.filtfilt(b, a, x_noise)
```

```python
plt.plot(t, x, color='k')
plt.plot(t, z, 'r--', t, z2, 'r:', t, y, 'r')
```

A common term in signals processing is called a __Transfer Function__, which basically is something that relates the output to the input.

$$\text{Output}(t) = h(\text{Input}(t))$$

This is horribly general and not interesting, but generally for LTI systems a common transfer function is multiplying in Fourier (or Laplace) space

$$\mathcal{F}[\text{Output}](\omega) = H(\omega) \mathcal{F}[\text{Input}](\omega)$$

Given from the, this means that the output is given by some convolution

$$ \text{Output}(t) = h(t) * \text{Input}(t) $$
